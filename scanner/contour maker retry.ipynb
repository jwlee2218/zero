```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import copy
```

```python
image= cv2.imread('frame.jpg')
```

```python
blurred = cv2.pyrMeanShiftFiltering(image, 31,91)
plt.imshow(blurred)
# cv2.imshow('blu',blurred)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
gray = cv2.cvtColor(blurred, cv2.COLOR_BGR2GRAY)
ret, threshold = cv2.threshold(gray, 0,255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
```

```python
_,contours,_ = cv2.findContours(threshold, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
print ("num of contours = %d " %(len(contours)))
```

```python
cv2.drawContours(image, contours, -1,(0,0,255),6)
plt.imshow(image)

cv2.imshow('_',image)
cv2.namedWindow('_',cv2.WINDOW_NORMAL)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
height = 1080
width = 1920
blank=np.zeros((height,width,3),np.uint8)
cv2.drawContours(blank, contours, -1, (255,255,255),6)
plt.imshow(blank)
```

```python
c = max(contours, key = cv2.contourArea)
```

```python
blank=np.zeros((height,width,3),np.uint8)
cv2.drawContours(blank, c, -1, (255,255,255),6)
plt.imshow(blank)
```

```python
blank=np.zeros((height,width,3),np.uint8)
rect = cv2.minAreaRect(c)
box1 = cv2.boxPoints(rect)
box = np.int0(box1)
im = cv2.drawContours(blank,[box],0,(255,255,255),6)
plt.imshow(im)
```
