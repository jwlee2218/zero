```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
import copy

```

```python
my_dpi = 96

x=[1,2,3,4]
y=[10, 20, 30, 40]
plt.figure(figsize=(800/my_dpi, 800/my_dpi), dpi=my_dpi)
plt.axis('off')
plt.plot(x,y)
plt.savefig('graph.png',bbox_inches='tight')
```

```python
graph=cv2.imread('graph.png')
graph=cv2.cvtColor(graph,cv2.COLOR_BGR2RGB)

plt.figure(figsize=(800/my_dpi, 800/my_dpi), dpi=my_dpi)
plt.axis('off')
plt.savefig('graph2.png',bbox_inches='tight')
plt.imshow(graph)
```

```python
graph2=cv2.imread('graph2.png')
graph2=cv2.cvtColor(graph,cv2.COLOR_BGR2RGB)

plt.figure(figsize=(800/my_dpi, 800/my_dpi), dpi=my_dpi)
plt.subplot(121),plt.imshow(graph)
plt.subplot(122),plt.imshow(graph2)
```
