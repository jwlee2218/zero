```python
import cv2

img1 = cv2.imread('./day2_lane1/220.jpg')
img2 = cv2.imread('./day2_lane1/221.jpg')
img3 = cv2.imread('./day2_lane1/222.jpg')

height , width , layers =  img1.shape

fourcc = cv2.VideoWriter_fourcc(*'mp4v') # Be sure to use lower case
video = cv2.VideoWriter('video.avi',fourcc,10.0,(width,height))

video.write(img1)
video.write(img2)
video.write(img3)

cv2.destroyAllWindows()
video.release()
```
